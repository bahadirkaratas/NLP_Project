```python
#articles = articles[0:759]
articles = articles[800:1000]
```

```python
truncated = []
for article in articles:
    splitted = article.split(" ")
    truncated.append(splitted[0:200])
```

```python
len(articles)
```

```python
truncated_text = []
for trunc in truncated:
    truncated_text.append(" ".join(trunc))
```

```python
import requests
import json


def generate_data_turkish(subject):
    URL = "https://api.openai.com/v1/chat/completions"

    payload = {
    "model": "gpt-3.5-turbo",
    "messages": [{"role": "user", "content": f"Aşağıdaki metni tam 200 kelime kullanarak Türkçe dilinde tekrar yazar mısın: {subject}"}],
    }

    headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer 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"
    }

    response = requests.post(URL, headers=headers, json=payload, stream=False)
    response.content
    print(response.content)
    

    data = json.loads((response.content).decode('utf-8'))
    text = (data['choices'][0]['message']['content'])
    f = open("train_chatgpt_rewrites_in_Turkish.txt", "a")
    f.write("\n")
    f.write(str(text))
    f.write("\n---------- o ----------")
    f.close()
def generate_data_english(subject):
    URL = "https://api.openai.com/v1/chat/completions"

    payload = {
    "model": "gpt-3.5-turbo",
    "messages": [{"role": "user", "content": f"Can you rewrite this text in English using 200 words: {subject}"}],
    }

    headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer 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"
    }

    response = requests.post(URL, headers=headers, json=payload, stream=False)
    response.content
    print(response.content)
    

    data = json.loads((response.content).decode('utf-8'))
    text = (data['choices'][0]['message']['content'])
    f = open("train_chatgpt_rewrites_in_English.txt", "a")
    f.write("\n")
    f.write(str(text))
    f.write("\n---------- o ----------")
    f.close()
```

```python
import time
for text in truncated_text[168:]:
    generate_data_turkish(text)
    time.sleep(20)
```
